## Scatter Plot

Verilerimize bakarak metre kareye göre fiyat tahmini yapmamız mümkün olur mu?


```python
import numpy as np
import matplotlib.pyplot as plt

plt.rcParams['figure.figsize'] = [30, 30]

fig, ax = plt.subplots()

ax.scatter(evler['LotArea'], evler['SalePrice'], alpha=0.2)


ax.set(xlabel='alan', ylabel='fiyat',
       title='fiyat/alan oranı')

plt.show()
```


Teker Teker zor oluyor. Bir döngüye alalım da hepsini göstersin bize

```python
plt.rcParams['figure.figsize'] = [10, 10]
for e in range(1,evler3.NeighborhoodID.max()+1):
    ax3 = evler3[evler3.NeighborhoodID ==e].plot.scatter(x='LotArea', y='SalePrice', c='OverallCond', s=size,alpha=0.5,colormap='viridis')
    ax3.legend(str(e))
    
    plt.show()
```
