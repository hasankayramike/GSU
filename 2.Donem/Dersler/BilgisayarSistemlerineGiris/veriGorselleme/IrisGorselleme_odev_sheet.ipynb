```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

import seaborn as sns
sns.set_style('whitegrid')
```

Iris veri setini, sklearn kütüphanesinin load iris komutu ile indirebiliyoruz.

```python
from sklearn.datasets import load_iris
iris=load_iris() # iris veri setini aldik.

print(iris)
```

Iris veri setinin data bölümlerine bakalım, bu çiçeklerin boyut bilgilerini tutuyor.

```python
print(iris['data'])
```

Feature names bolumu, data daki her bir sutunun ne ise yaradigi bilgisini tutuyor.

```python
print(iris['feature_names'])
```

Target bolumu, bu ciceklerin turunu bize soyluyor.

```python
print(iris['target'])
```

Target names, 0, 1 ve 2 degerlerinin ne oldugunu bize bildiriyor.

```python
print(iris['target_names'])
```

Veri setimizi pandas ile işleyebilmek için bunu bir pandas data frameine çevirelim.

```python
data = pd.DataFrame(iris['data'], columns=iris['feature_names'])
data['target'] = iris['target']   # BURASI BILINEN NETICELERI ICERIYOR
```

# Veri Gorselleme

Veri setimizi görselleyelim. Verimize daha aşina oluruz.

2. Her bir çiçek türünden kaç adet olduğunu bir "pie chart" üzerinde gösteriniz.

```python
# pie chart
labels = iris["target_names"]
label_codes = iris["target"]

element, count = np.unique(label_codes, return_counts=True)
unique_dict = dict(zip(element, count))
sizes = list(unique_dict.values())

fig1, ax1 = plt.subplots()
ax1.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=90)
plt.title('Çiçek Türlerinin Dağılımı')
plt.show()
```

Her bir tür için bulunan veri miktarı hakkında ne diyebiliriz?

# Verilerin incelenmesi

3. Verimizin her bir sütunu için histogram çıkarınız.

```python
num_bins = 25
for i in iris["feature_names"]:
    plt.hist(data[i], num_bins, density=1, facecolor='blue', alpha=0.5)
    plt.xlabel(i)
    plt.show()
```

4. Verimizdeki her bir sütün için boxplot oluşturun.
https://www.geeksforgeeks.org/box-plot-in-python-using-matplotlib/

```python
for i in iris["feature_names"]:
    plt.boxplot(data[i])
    plt.show()
```

5. Sepal ve petal uzunluklarını ikişer ikişer birer scatter plot ile görselleyin.

```python
tmp1 = list(data["sepal length (cm)"])
tmp2 = list(data["petal length (cm)"])
tmp3 = list(iris["target"])

sepal = tmp1[::2]
petal = tmp2[::2]
x_axis = tmp3[::2]

plt.scatter(x_axis, sepal, alpha=0.3)
plt.xlabel("Target")
plt.ylabel("Sepal Length (cm)")
plt.show()

plt.scatter(x_axis, petal, alpha=0.3, cmap='viridis')
plt.xlabel("Target")
plt.ylabel("Petal Length (cm)")
plt.show()
```
